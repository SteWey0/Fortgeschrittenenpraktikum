```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
plt.rcParams.update({'font.size': 16, 'lines.markersize': 8})
cmap = get_cmap('tab10').colors
```

```python
def fermi(x,ef,T):
    return 1/(np.exp((x-ef)/(T*8.617e-5))+1)
e = np.linspace(-1,1,10000)
fig,ax = plt.subplots(figsize=(7.5,5))
ax.plot(e,fermi(e,0,1000)*100,label='T=1000K',color=cmap[0])
ax.plot(e,fermi(e,0,500)*100,label='T=500K',color=cmap[1])
ax.plot(e,fermi(e,0,0)*100,label='T=0K',color=cmap[2])
ax.legend()
ax.grid()
ax.set_xlabel('Abweichung von der Fermi-Energie E-$\epsilon_F$ [eV]')
ax.set_ylabel('Besetzungsgrad [%]')
ax.set_xticks(np.arange(-1,1.1,0.5))
ax.set_yticks(np.arange(0,120,25))
fig.tight_layout()
fig.savefig('fermi.pdf')
plt.show()
```
