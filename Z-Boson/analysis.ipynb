```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('../fullwidth.mplstyle')
```

```python
bhabha_s = lambda x : 1 + (x)**2
bhabha_t = lambda x : (1 - x)**-2
# x = np.cos(np.deg2rad((np.linspace(1, 179, 1000))))
x = np.linspace(-1,1, 1000)
fig,ax = plt.subplots()
ax.plot(x, bhabha_s(x) + bhabha_t(x), label='total', color='black')
ax.plot(x, bhabha_s(x), label='s channel')
ax.plot(x, bhabha_t(x), label='t channel')
ax.set_xlabel('Cosine of scattering angle')
ax.set_ylabel('Differential cross section [a.u.]')
ax.set_yscale('log')
ax.legend()
fig.tight_layout()
fig.savefig('plots/theory_bhabha_cos.pdf', bbox_inches='tight')
```
