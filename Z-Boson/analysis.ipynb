```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('../fullwidth.mplstyle')
```

```python
bhabha_s = lambda x : 1 + (np.cos(np.deg2rad(x)))**2
bhabha_t = lambda x : (1 - np.cos(np.deg2rad(x)))**-2
x = np.linspace(1, 179, 1000)
fig,ax = plt.subplots()
ax.plot(x, bhabha_s(x) + bhabha_t(x), label='total', color='black')
ax.plot(x, bhabha_s(x), label='s channel')
ax.plot(x, bhabha_t(x), label='t channel')
ax.set_xlabel('Angle [°]')
ax.set_ylabel('Differential cross section [a.u.]')
ax.set_yscale('log')
ax.legend()
fig.tight_layout()
fig.savefig('plots/theory_bhabha.pdf', bbox_inches='tight')
```
